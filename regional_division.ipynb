```python
# 邻接矩阵约束
def neighbor(ch, neighbor_m):
    x = ch.copy()
    a = neighbor_m.copy()
    c = np.matmul(x, a)
    cc = np.multiply(c,x)
    k = 0
    for i in range(m):
        for j in range(n):
            if cc[i][j] == 1:
                k = k-1
    return k

print(neighbor(chrome, neighbor_metric))            
    

        
```

```python
def repeat(solutions, solutions_list):
    solutions_l = np.array(solutions_list)
    b = 0
    for i in range(len(solutions_list)):
        if ((solutions == solutions_l[i]).all()) == True and len(solutions_l[i]) != 20:
            b = 1
    return b           


def init_solutions(city, serve_position, neighbor_metric):
    k = 0
    
    solutions_list = []
    while k < 20:
        a = 0
        solutions = np.zeros((len(serve_position),city),dtype=int)
        for i in range(city):
            if i == 1 or i == 8 or i == 12 or i==18:# i == 1 or i == 8 or i == 12 or i==18
                solutions[a][i] = 1
                a = a+1
            else:               
                j = np.random.choice(serve_center)
                solutions[j][i] = 1
        if neighbor(solutions, neighbor_metric) == 0 and repeat(solutions, solutions_list) == 0:
            k = k + 1
            solutions_list.append(solutions)
    return solutions_list
solutions_list = init_solutions(city, serve_position, neighbor_metric)
print(solutions_list)
```
